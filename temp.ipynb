```sage
from sage.all import *

x = var("x")
```

```sage
phi = x / log(x)
diff_phi = diff(phi, x) 
diff_phi
```

```sage
# Define the functions and their plots
plot1 = plot(log(x), (x, 2, 100), color="blue")
plot2 = plot(phi, (x, 2, 100), color="black")  # replace phi with your function
plot3 = plot(diff_phi, (x, 2, 100), color="green")  # replace diff_phi with your function

# Combine the plots
combined_plot = plot1 + plot2 + plot3

# Define the y-axis limits
ymin = 0  # replace with your desired minimum y value
ymax = 10  # replace with your desired maximum y value

# Show the plot with the specified y-axis limits
combined_plot.show(ymin=ymin, ymax=ymax)
```
