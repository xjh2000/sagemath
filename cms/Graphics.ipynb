```python
from operator import le


def p(x, n):
    return taylor(sin(x), x, 0, n)


xmax = pi
n = 3
g = plot(sin(x), x, -xmax, xmax, legend_label="sin(x)")
for d in range(n):
    g += plot(
        p(x, 2 * d + 1),
        x,
        -xmax,
        xmax,
        color=(1.7 * d / (2 * n), 1.5 * d / (2 * n), 1 - 3 * d / (4 * n)),
        legend_label="p(x, %d)" % (2 * d + 1),
    )
g.show(ymin=-2, ymax=2)
```

```python
n, l, x, y = 10000, 1, 0, 0
p = [[0, 0]]
for k in range(n):
    theta = (2 * pi * random()).n(digits=5)
    x, y = x + l * cos(theta), y + l * sin(theta)
    p.append([x, y])
g1 = line([p[n], [0, 0]], color="red", thickness=2)
g1 += line(p, thickness=0.4)
g1.show(aspect_ratio=1)
```

```python
length = 10000
n = var("n")
u = lambda n:  n *ln(n) * sqrt(2)
z = lambda n: exp(2 * I * pi * u(n)).n()
vertices = [CC(0, 0)]
for n in range(1, length):
    vertices.append(vertices[n - 1] + CC(z(n)))
line(vertices).show(aspect_ratio=1)
```

```python
x = var("x")
y = function("y")
DE = x * diff(y(x), x) == 2 * y(x) + x ^ 3
desolve(DE, [y(x), x])
sol = []
for i in srange(-2, 2, 0.2):
    sol.append(desolve(DE, [y(x), x], ics=[1, i]))
    sol.append(desolve(DE, [y(x), x], ics=[-1, i]))
g = plot(sol, x, -2, 2)
y = var("y")
g += plot_vector_field((x, 2 * y + x ^ 3), (x, -2, 2), (y, -1, 1))
g.show(ymin=-1, ymax=1)
```
