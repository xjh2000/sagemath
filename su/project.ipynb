## Computing a Selling Price

$$x=360-80p$$
$$p=4.5-\frac{x}{80}$$
$$r=xp=4.5x-\frac{x^2}{80}$$
$$c=0.5x+100$$
$$profit=-\frac{x^2}{80}+4n-100$$

$x$ is the number of units sold, $p$ is the price per unit, $r$ is the revenue, $c$ is the cost, and $profit$ is the profit. 

```python
x = var("x")
r = 4.50 * x - x ^ 2 / 80
c = 0.5 * x + 100
p = r - c

mp_x = find_root(diff(p), 0, 360)
mp_y = p(x=mp_x)

mp = point((mp_x, mp_y), color="red", size=30)
P1 = plot(r, 0, 360, color="green")
P2 = plot(c, 0, 360, color="red")
P3 = plot(p, 0, 360, color="black", linestyle="--")

P = P1 + P2 + P3 + mp
P.show()
```
