```python
z, phi = var('z, phi')
eq = z**2 - 2/cos(phi)*z + 5/cos(phi)**2 - 4 == 0
latex(solve(eq, z))
```

$$z^{2} - \frac{2 \, z}{\cos\left(\phi\right)} + \frac{5}{\cos\left(\phi\right)^{2}} - 4 = 0 \\
\left[z = -\frac{2 \, \sqrt{\cos\left(\phi\right)^{2} - 1} - 1}{\cos\left(\phi\right)}, z = \frac{2 \, \sqrt{\cos\left(\phi\right)^{2} - 1} + 1}{\cos\left(\phi\right)}\right]
$$
